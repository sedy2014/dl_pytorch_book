```python
from PIL import Image

# Load the image using PIL
img = Image.open('../data/p2ch10/astronaut.png')
# Define a transform to convert the image to a tensor
transform = transforms.Compose([
    transforms.ToTensor(),
])
# Apply the transform to the image
tensor = transform(img)
noisy_tensor, noise = forward_diffusion_sample(tensor, 150)

# Convert the noisy tensor back to a PIL image
noisy_img = transforms.ToPILImage()(noisy_tensor)

width, height = img.size
new_width = int(width * 0.5)
new_height = int(height * 0.5)
# Resize the image
noisy_img = noisy_img.resize((new_width, new_height))

noisy_img
```
